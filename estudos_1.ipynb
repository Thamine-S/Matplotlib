```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Gráfico básico

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

#Mudar o tamanho da figura
plt.figure(figsize= (5,3), dpi = 300) # pixel por polegada

plt.plot(x, y, label='linha', color = "purple", linewidth= 2, marker = '.', markersize = 10, markeredgecolor = 'black', linestyle = '--' ) # é possível escolher uma cor específica atravez de um código hex 
# fmt = '[color][marker][line]'
#plt.plot(x,y, 'b^--', label='linha')

#Segunda Linha

# Selecionando um intervalo que queremos colocar pontos
x2 = np.arange(0, 4.5, 0.5)

#dividindo parte da linha em cor sólida e pontilhado
plt.plot(x2[:5], x2[:5]**2, label='linha 2', color = 'pink')
plt.plot(x2[4:], x2[4:]**2, color = 'pink',linestyle ='--' )

plt.title('Meu Primeiro Gráfico!', fontdict={'fontsize': 20}) # utilizando o fontdict é possível alterar muitos dos recursos escritos do gráfico

plt.xlabel('Eixo X!')
plt.ylabel('Eixo Y!')

# O gráfico se reajusta de acordo com os ticks que escolhemos, portanto é possível fazer muitas distorções
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10, 14, 18, 22])

plt.legend()

plt.savefig('graph.png', dpi=300)

plt.show() #isso retira a linha em cima do código, é a maneira certa de criar um gráfico
```

# Gráfico de barras

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars=plt.bar(labels, values)

#existem várias formas de estilizar um gráfico de barras:

#forma 1
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

#forma 2
paterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(paterns.pop())

plt.figure(figsize = (6,4), dpi = 300)

#Nesse exemplo não foram criadas legendas ou um título para o gráfico



plt.show()
```

## Gráficos de Pizza

### Gráfico de pizza simples

```python

# fifa.loc -> encontra determinada informação dentro da BD
# A  informação que queremos é a quantidade de jogadores que preferem o pé esquerdo
# Portanto usa-se o .count() que nos retorna um arrya, porém junto do [0] ele nos retorna a soma de todos os jogadores que preferem a perda esquerda


left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]


print(f""" 
Quantidade de jogadores que preferem o pé esquerdo: {left}
Quantidade de jogadores que preferem o pé direito: {right}
""")
```

```python
#criação do gráfico em Pizza
labels = ['Pé esquerdo', 'Pé direito']
cor = ['#aabbcc', '#abcdff']

plt.pie([left, right], labels=labels, colors=cor, autopct = '%.2f %%')

plt.title('Preferência de pé dos jogadores da fifa (2018)')


plt.show()
```

#### Limpeza de dados
Com base na lista da coluna de 'Weight' podemos perceber que os dados estão em libras, e estão identificados como object, por possuir "lbs" no final do número. 

Portanto iremos:
- Converter de libras para Quilogramas
- Retirar o 'lbs' 
- Mudar o tipo de object para Int

```python
#Retirando o sufixo 'lbs' e convertendo para int:
# If case pro extenso:
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
# Convertendo para quilogramas e formatando para 2 casas após a virgula
#fifa['Weight'] = fifa['Weight']* 0.453592
#fifa['Weight'] = fifa['Weight'].round(2)

fifa.Weight
```

```python
light = fifa.loc[fifa.Weight < 54]. count()[0]
light_medium = fifa.loc[(fifa.Weight >= 54) & (fifa.Weight < 68)].count()[0]
medium = fifa.loc[(fifa.Weight >= 68) & (fifa.Weight < 79)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 79) & (fifa.Weight < 90)].count()[0]
heavy = fifa.loc[fifa.Weight > 90].count()[0]

pesos = f'''
Quantidade de jogadores na categoria peso leve: {light}
Quantidade de jogadores na categoria peso médio Leve: {light_medium}
Quantidade de jogadores na categoria peso médio: {medium}
Quantidade de jogadores na categoria peso médio pesado: {medium_heavy}
Quantidade de jogadores na categoria peso Pesado: {heavy}
'''
print(pesos)
```

```python
#criação do gráfico em Pizza
labels = ['Peso Leve: abaixo de 54kg', 'Peso Médio-Leve: 54-68kg', 'Peso Médio: 68-79kg', 'Peso Médio-Pesado: 79-90kg', 'Peso Pesado: acima de 90kg']
cor = ['#00021f', '#8123a1', '#a18e23', '#a13023','#2390a1']

explode = (.4,.1,.1,.1,.4)
#explode é uma propriedade que basicamente
plt.pie([light, light_medium, medium, medium_heavy, heavy], colors=cor, autopct = '%.2f %%', pctdistance=1.2, explode = explode)

plt.title('Porcentagem de jogadores da FIFA por categoria de peso (2018)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', labels= labels)


plt.show()
```

## Gráfico diagrama de caixa

### Comparando times

```python
plt.style.use('default')
plt.figure(figsize=(10,15))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
r_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
psg = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']
manch_Uni = fifa.loc[fifa.Club == 'Manchester United']['Overall']
manch_Cit = fifa.loc[fifa.Club == 'Manchester City']['Overall']
chealsea = fifa.loc[fifa.Club == 'Chelsea']['Overall']
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
a_madrid = fifa.loc[fifa.Club == 'Atlético Madrid']['Overall']

labels = ['Barcelona', 'Real\nMadrid', 'Paris\nSaint-Germain', 'Manchester\nUnited', 'Manchester\nCity', 'Chelsea', 'Juventus', 'Atlético\nMadrid']

boxes = plt.boxplot([barcelona, r_madrid, psg, manch_Uni, manch_Cit, chealsea, juventus, a_madrid], labels=labels, patch_artist=True, medianprops = {'linewidth': 2})

for box in boxes['boxes']:
    box.set(color= '#7070d4', linewidth=2)
    
    box.set(facecolor= '#c7c7ed')

plt.legend()

plt.title('Comparação de Times FIFA (2018)')
plt.ylabel('FIFA hanking geral')

plt.show()
```
